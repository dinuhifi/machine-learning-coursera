# Implementing linear regression using only numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,5,10])
y_train = np.array([3,7,12])
```

```python
plt.plot(x_train,y_train)
```

```python
w = 0
b = 0
```

```python
import itertools
def cost(x_train,y_train,w,b):
    c = 0
    for i,j in zip(x_train,y_train):
        c = ((i*w + b) - j)**2
    c = c/6
    return c
```

```python
cost(x_train,y_train,w,b)
```

```python
def compute_gradient(x, y, w, b): 
    m = x.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):  
        f_wb = w * x[i] + b 
        dj_dw_i = (f_wb - y[i]) * x[i] 
        dj_db_i = f_wb - y[i] 
        dj_db += dj_db_i
        dj_dw += dj_dw_i 
    dj_dw = dj_dw / m 
    dj_db = dj_db / m 
        
    return dj_dw, dj_db
```

```python
compute_gradient(x_train,y_train,w,b)
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,iters,grad):
    b = b_in
    w = w_in
    for i in range(iters):
        dj_dw, dj_db = grad(x, y, w , b)     
        b = b - alpha * dj_db                            
        w = w - alpha * dj_dw

        if (i+1)%100 == 0:
            print(f'iteration:{i+1}: cost={cost(x,y,w,b)}')
        
    return w,b
```

```python
pred = gradient_descent(x_train,y_train,w,b,0.01,300,compute_gradient)
pred
```

```python
plt.plot(x_train, pred[0]*x_train + pred[1],"b--", label="Prediction")
plt.plot(x_train,y_train,'g--',label="Original")
plt.legend()
```
